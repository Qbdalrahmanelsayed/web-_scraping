```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


```

```python
page =requests.get("https://books.toscrape.com/")
src =page.content
```

```python
soup =BeautifulSoup(src,"html.parser")
soup
```

```python
book_list=[]
for book in soup.find_all("li",class_="col-xs-6 col-sm-4 col-md-3 col-lg-3"):
  
    img='https://books.toscrape.com/' +book.find("img" ,class_="thumbnail")["src"]
    title=book.find("h3").text
    price=book.find("p" ,{"class":"price_color"}).text.strip()
    rating=book.find("p" ,{"class":"star-rating Three"})
    book_list.append([title , price ,rating ,img])
    
for i in book_list:
    print(i)
```

```python
df = pd.DataFrame(book_list,columns=["title" ," price" ,"rating" , "img"])
df
```

```python
df.to_csv("book.csv")
```
